## 클래스가 필요한 이유

계산기에 숫자 3을 입력하고 +기호를 입력하고 4를 입력하면 결과적으로 7을 보여준다.

즉, 3이라는 숫자를 메모리 어딘가에 저장을 하고 있어야 함

```python
result = 0

def add(num):
    global result # 외부의 result 변수를 함수 내에서 활용
    result += num
    return result
```

```python
add(3)
add(4)
```

```python
# 만약 한 프로그램에서 2대의 계산기가 필요하다면?
result1 = 0
result2 = 0

def add1(num):
    global result1
    result1 += num
    return result1

def add2(num):
    global result2
    result2 += num
    return result2
```

```python
# add1의 계산기
print(add1(3))
print(add1(4))
```

```python
# add2의 계산기
print(add2(3))
print(add2(7))
```

```python
## 클래시 예시
class Calculator:
    def __init__(self):
        self.result = 0
    
    def add(self, num):
        self.result += num 
        return self.result
```

```python
cal1 = Calculator() # cal1이라는 객체 생성
cal2 = Calculator() # cal2라는 객체 생성
```

```python
print(cal1.add(3))
print(cal1.add(4))
```

```python
print(cal2.add(3))
print(cal2.add(7))
```

## 생성자(Constructor)

```python
a = Fourcal()
a.add()
```

a.set_data 메서드를 수행하지 않고 add를 수행하면 error가 발생, set_data 메서드를 수행해야만 함
초기값을 매 번 생성 혹은 설정하기보다는 생성자를 직접 구현하는 것이 편리하고 안전하다

생성자(Constructor)란 객체가 생성될 때 자동으로 호출되는 메서드를 의미

```python
# 객체에 생성자를 추가해주기-4
class Fourcal:
    ## 메서드 method - 클래스 내에서 존재하는 함수
    # __init__함수는 생성자임
    def __init__(self, first, second):
        self.first = first
        self.second = second    
    def set_data(self, first, second): # 메서드 self, first, second 메서드의 매개변수(parameter)
        self.first = first
        self.second = second  
    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
```

```python
# __init__이라는 메서드는 set_data와 이름만 다르고 모든것이 동일함. 단, 메서드 이름을 __init__했기에
# 생성자로 인식되어 객체가 생성되는 시점에 자동적으로 호출이 됨
```

```python
## init함수를 추가하고 나면 
a = Fourcal(3,5)
a.add()
a.mul()
```

### 클래스의 상속(inheritance)

만약에 Fourcal 클래스에 a의 b제곱이라고 하는 기능을 추가 혹은 첨가한다면 어떻게 해야될까 ? 

```python
# Fourcal에 해당하는 기능들을
# 그대로 powfourcal에도 적용하는 것

class powfourcal(Fourcal):
    pass
```

```python
a = powfourcal(4,3)
a.add()
```

    * 왜 상속을 해야되지?
    
    1) 보통 상속은 기존 클래스를 변형하지 않고 기능을 추가하거나, 일부 기능의 변경을 위해 사용됨
    2) 또한, 기존 클래스가 라이브러리 형태로 제공되거나 수정이 허용되지 않는 상황이라면 상속을 사용
    

```python
# 이제 a의 b제곱을 구현해보기
class powfourcal(Fourcal):
    def pow1(self):
        result = self.first ** self.second
        return result
```

```python
a1 = powfourcal(3,2)
a1.pow1()
```

### 메서드 오버라이딩

```python
a = Fourcal(4,0)
a.div() # 만약에 error값을 호출하지 않고 
        # 0이라는 keyword값을 반환하고 싶다면?
        # 어떻게 해야할까?
```

```python
# 상속을 활용하여 divfourcal 생성
class divfourcal(Fourcal):
    # Fourcal 클래스에 있는 div메서드(클래스내의 함수에)
    # 같은 이름으로 또 div를 활용하였음 ## 메서드 오버라이딩(overriding, 덮어쓰기)
    def div(self):
        if self.second == 0: # 나누는 값이 0이라면, 그냥 0을 리턴
            return 0
        else:
            return self.first/self.second           
```

```python
test = divfourcal(5,0)
test.div()
```

## Q2 객체변수가 100이상의 값을 가질 수 없도록 하는 limited_cal 클래스를 만들어보자.

```python
cal = limited_cal()
cal.add(50)
cal.add(60)

print(cal.value) ## 이 값은 100이 출력되도록
```

```python
# 단, 반드시 cal 클래스를 상속하여 만드세요
class cal:
    def __init__(self):
        self.value = 0
        
    def add(self,val):
        self.value += val 
```

```python
class limited_cal(cal):
    def add(self,val):
        self.value +=val
        if self.value >100:
            self.value = 100
```

```python
cal = limited_cal()
cal.add(50)
cal.add(60)

print(cal.value) ## 이 값은 100이 출력되도록
```
