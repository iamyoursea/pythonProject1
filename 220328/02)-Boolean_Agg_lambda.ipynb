### groupby 적용

```python
titan_groupby = titan_df.groupby('Pclass')[['Cabin','Age','Name']].count()
titan_groupby
```

```python
titan_df.groupby('Sex')[['Cabin','Age','Name']].count()
```

```python
# titan_df에서 pclass를 기준으로 각 객실등급별로 Age와 Fare의 max와 min값을 보고 싶을때?
titan_df.groupby('Pclass')[['Age','Fare']].agg(['min','mean','max'])
```

```python
# 만약 위와 같은 상황에서 ; Age는 max, Sibsp은 sum, Fare는 mean을
# 나타내고 싶다면???

agg_form = {'Age':'max', 'SibSp':'sum', 'Fare':'mean'}
titan_df.groupby('Pclass').agg(agg_form)
```

## apply lambda 데이터 가공

```python
# Name의 길이를 구하는 lambda식 적용
```

```python
titan_df['Na_len'] = titan_df['Name'].apply(lambda x: len(x))
titan_df.head()
```

```python
# 조금 더 복잡한 if else절을 활용한 가공
# 나이가 18세 미만 :: 'child' OR 'adult' 
print(titan_df['Age'].apply(lambda x: 'young' if x < 18 else 'adult').value_counts())
# titan_df['Age_cat'] = titan_df['Age'].apply(lambda x: 'young' if x < 18 else 'adult')
titan_df[titan_df.Age_cat=='young'].shape
```

```python
# 조금 더 복잡한 if else절을 활용한 가공
# 나이가 18세 미만 :: 'child' 60세미만 :: 'adult' OR 'elderly'

print(titan_df['Age'].apply(lambda x: 'young' if x < 18 else ('adult' if x<60 else 'elderly')).value_counts())
```

```python
def get_category(age):
    cat=''
    if age <=5: cat='Baby'
    elif age<=12: cat= 'Child'
    elif age<=18: cat='Teenager'
    elif age<=25: cat='Student'
    elif age<=35: cat='Young Adult'
    elif age<=60: cat= 'Adult'
    else: cat='Elderly'
    
    return cat
```

```python
titan_df.Age.apply(lambda x : get_category(x))
```
